```python
info_df = pd.DataFrame(columns=['exercise', '#submissions','errored', 'passed', 'failed', 'passed_with_warnings', '#submissions_to_passed', 'students'])

for leccion in guias:
    for exercise in leccion:
        short_submissions[(short_submissions['exercise.name'] == exercise) & (short_submissions['status'] == 'passed')] = short_submissions[(short_submissions['exercise.name'] == exercise) & (short_submissions['status'] == 'passed')].drop_duplicates('submitter.email', keep='last')
        exer_df = short_submissions[short_submissions['exercise.name'] == exercise]
        submissions_grouped = exer_df.groupby(['status']).size()
        total_amount_submissions = exer_df.shape[0]
        metrics = {}
        metrics = submissions_grouped
        metrics['exercise'] = exercise
        metrics['#submissions'] = total_amount_submissions
        metrics['students'] = exer_df['submitter.email'].nunique()
        metrics['#submissions_to_passed'] = short_submissions[(short_submissions['exercise.name'] == exercise) & (short_submissions['status'] == 'passed')]['submissions_count'].mean()
        info_df = info_df.append(metrics, ignore_index=True)
```
